```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-4,4,25)
y=np.linspace(-4,4,25)
X,Y=np.meshgrid(x,y)
```

```python
def Flujo(x,y,R):
    V=2
    return (V*x*(1-(R**2)/(x**2+y**2)))
```

```python
def dxFlujo(x,y,h,R):
    return (Flujo(x+1,y,R)-Flujo(x-1,y,R))/(2*h)
def dyFlujo(x,y,h,R):
    return -1*(Flujo(x,y+1,R)-Flujo(x,y-1,R))/(2*h)
VX=np.zeros_like(X)
VY=np.zeros_like(Y)
h=0.01
R=2
for i in range(25):
    for j in range(25):
        if (X[i,j]**2+Y[i,j]**2)>=R**2:
            VX[i,j]=dxFlujo(X[i,j],Y[i,j],h,R)
            VY[i,j]=dyFlujo(X[i,j],Y[i,j],h,R)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
circle=plt.Circle((0,0),2,color='b')
ax.add_patch(circle)
for i in range(25):
    for j in range(25):
        ax.quiver(x[i],y[j],VX[i,j],VY[i,j])
```
