```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
def cumpleaños (n):
    p1= 364/365 #probabilidad de que el dia elegido no sea el del cumpleaños
    if n ==1:
        return p1
    else:
        b=np.linspace(1,n,n)
        k=2
        a = factorial(n) / (factorial(k) * factorial(n - k))
    return p1**a
```

```python
x=np.linspace(1,80,80)
b= np.zeros_like(x)
for i in range(len(x)):
    b[i]=cumpleaños(int(x[i]))
plt.plot(x,b)
plt.ylabel('Probabilidad')
plt.xlabel('Numero de personas')
plt.show()
```
