```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
p1=np.linspace(0.1,0.9,1001)
p2=np.linspace(0.1,0.5,1001)
p1,p2=np.meshgrid(p1,p2)
```

```python
def P(p1,p2,p3=0.5,p4=0.5):
    a=p1*p2*(1-p3)*(1-p4)
    b=p1*(1-p2)*p3*(1-p4)
    c=p1*(1-p2)*(1-p3)*p4
    d=(1-p1)*p2*p3*(1-p4)
    e=(1-p1)*p2*(1-p3)*p4
    f=(1-p1)*(1-p2)*p3*p4
    return(a+b+c+d+e+f)
```

```python
PA=P(p1,p2)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(p1, p2, PA, cmap='viridis', edgecolor='none')
```

```python
PAplano=PA.flatten()
p1plano=p1.flatten()
p2plano=p2.flatten()
minimo=min(PA.flatten())
maximo=max(PA.flatten())
nminimo=np.where(PAplano==minimo)[0][0]
nmaximo=np.where(PAplano==maximo)[0][0]
```

```python
print("El minimo de probabilidad posible en estos intervalos es: ")
print(minimo)
print("y este mÃ­nimo se alcanza con las variables (p1,p2) iguales a")
print("("+str(p1plano[nminimo])+","+str(p2plano[nminimo])+")")
print("")
print("El mÃ¡ximo de probabilidad posible en estos intervalos es: ")
print(maximo)
print("y este mÃ¡ximo se alcanza con las variables (p1,p2) iguales a")
print("("+str(p1plano[nmaximo])+","+str(p2plano[nmaximo])+")")
```
