```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(np.tan(x))
```

```python
h=0.01
x=np.arange(0.1,1.1,h)
```

```python
def dprog(x,f,h):
    return (-3*f(x)+4*f(x+h)-f(x+2*h))/(2*h)
```

```python
def dcentral(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
def dexacta(x):
    return 1/(2*np.sqrt(np.tan(x))*np.cos(x)**2)
```

```python
yp=dprog(x,f,h)
yc=dcentral(x,f,h)
ye=dexacta(x)
```

```python
plt.plot(x,yp)
plt.plot(x,yp)
plt.plot(x,ye)
```

```python
errorp=yp-ye
errorc=yc-ye
```

```python
plt.scatter(x,errorp,s=1)
plt.scatter(x,errorc,s=1)
plt.axhline(0,color='k',linestyle='dashed')
```

Aparentemente sí tienen el mismo orden de error (claro aunque uno es positivo y el otro negativo), pero la similaridad en las lineas de error puede apoyar esta teoría.
