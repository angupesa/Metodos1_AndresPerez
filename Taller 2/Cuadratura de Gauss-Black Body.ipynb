In the black-body radiation problem the following integral appears:
    $$\int_{0}^{\infty} \frac{x^3}{e^x-1}$$
    
a)Compute this integral using the Gauss-Laguerre quadrature method for n=3 evaluation
points.


b) For this estimation, plot the relative error (r(n) = Iestimated(n)/Iexact) as a function
of the evaluation points, with n = [2, 3, ..., 10] [3.6].

```python
import numpy as np
import sympy as sym
from scipy import integrate
import matplotlib as mpl
import matplotlib.pyplot as plt

def f (x):
    return (np.exp(x)*(x**3))/((np.exp(x)**2-np.exp(x)))
def integrate(f,n):
  X,W=np.polynomial.laguerre.laggauss(n)
  inte = np.sum(f(X)*W*np.exp(X))
  return inte
a=integrate(f,3)
a
```

```python
n=np.array([2,3,4,5,6,7,8,9,10])
def error(f,n):
    exact= np.pi**4/15
    return integrate(f,n)/exact
y=np.array([error(f,n[0]),error(f,n[1]),error(f,n[2]),error(f,n[3]),error(f,n[4]),error(f,n[5]),error(f,n[6]),error(f,n[7]),error(f,n[8])])
```

```python
plt.scatter(n,y,label="Precision de los resultados de la interpolacion.")
plt.ylim([0.99, 1.001])
plt.xlabel("n")
plt.ylabel("Incertidumbre de interpolacion")
plt.legend()
```
